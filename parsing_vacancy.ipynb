```python
import parse_hh_data
from selenium.webdriver.common.by import By
from selenium import webdriver
import re
from parse_hh_data import download, parse
import csv
import time
from urllib.error import HTTPError
```

```python
RESUME = (By.XPATH, '//a[@data-qa="serp-item__title"]')
VACANCY = (By.XPATH, '//span/a[@class="bloko-link"]')
```

```python
driver = webdriver.Chrome()
id_resume = []
resumes = []

for i in range(0, 50):
    driver.get(f"https://spb.hh.ru/search/resume?items_on_page=100&page={i}")
    elements = driver.find_elements(*RESUME)
    for element in elements:
        try:
            stroka = element.get_attribute('href').split('/')[-1].split('?')[0]
            try:
                resume = parse.resume(download.resume(stroka))
                if resume is not None:
                    resumes.append(resume)
            except HTTPError:
                print('ошибка')
        except AttributeError as e:
            print(f"An AttributeError occurred: {e}")
driver.quit()
```

```python
import requests
def download_vacancy(vacancy_id):
    url = f"https://api.hh.ru/vacancies/{vacancy_id}"
    try:
        response = requests.get(url)
        if response.status_code == 200:
            return response.json()
        else:
            print(f"Failed to download vacancy {vacancy_id}. Status code: {response.status_code}")
            return None
    except Exception as e:
        print(f"Failed to download vacancy {vacancy_id}. Error: {str(e)}")
        return None

# Initialize WebDriver
driver = webdriver.Chrome()

# Your vacancy extraction logic here
id_vacancy = []
vacancies = []

for i in range(0, 40):
    driver.get(f"https://spb.hh.ru/search/vacancy?page={i}")
    elements = driver.find_elements(*VACANCY)
    for element in elements:
        vacancy_id = re.sub(r'&', r'?', re.sub(r'click\?b=', r'', element.get_attribute('href').split('/')[-1])).split('?')[0]
        vacancy_info = download_vacancy(vacancy_id)
        if vacancy_info:
            vacancies.append(vacancy_info)
        else:
            print(f"Vacancy {vacancy_id} skipped due to error.")
    time.sleep(1)  # Add delay to avoid being blocked

# Quit WebDriver
driver.quit()
```
